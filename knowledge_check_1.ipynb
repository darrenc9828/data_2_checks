# Import libraries


```python
# Import libraries

import requests
import pandas as pd
```

# Pull in data from an API

```python
url = 'https://openlibrary.org/subjects/programming.json'
api = requests.get(url)

```

```python
json = api.json()
print (json)
```

```python
df = pd.DataFrame(json)
df
```

```python
df.columns
```

# Clean data to see nested dictrionaries

```python
works = pd.json_normalize(df.works)
works
```

# Find and print TWO descriptive statistics

```python
works['title'].value_counts()
```

```python
contain_values = works[works['title'].str.contains('Co')]
contain_values
```
